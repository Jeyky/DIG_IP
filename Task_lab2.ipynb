```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_half=cv2.imread("./assignments/Lab3_half100.png")
img_chess=cv2.imread("./assignments/Lab3_chess100.png")
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img_half,cmap='gray')
plt.subplot(122)
hist_half = cv2.calcHist([img_half], [0], None, [256], [0, 256])
plt.plot(hist_half); plt.xlim([-5, 260]); plt.grid(); plt.show();

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img_chess)
plt.subplot(122)
hist_chess=cv2.calcHist(img_chess,[0],None,[256],[0, 256])
plt.plot(hist_chess);plt.xlim([-5, 260]); plt.grid(); plt.show();

#---------------EXTRA TASK--------------------------------------

img_wood=cv2.imread("./examples/Wood.png",cv2.IMREAD_COLOR)
img_lwood=cv2.imread("./examples/lwood.jpg",cv2.IMREAD_COLOR)
img_wood=cv2.cvtColor(img_wood,cv2.COLOR_BGR2RGB)
img_lwood=cv2.cvtColor(img_lwood,cv2.COLOR_BGR2RGB)
img_wood_crop=img_wood[400:450,400:450,:]
img_lwood_crop=img_lwood[400:450,400:450,:]
chess=np.zeros((400,400,3),dtype='uint8')
h=range(1,41)#number of squares
aoi=50;#area of interest
for i in range(0,int(chess.shape[0]/aoi)):
    for j in range(0,int(chess.shape[1]/aoi)):
        if h[i+j]%2!=0:
            chess[i*aoi:i*aoi+aoi,j*aoi:j*aoi+aoi,:]=img_lwood_crop
        else:
            chess[i*aoi:i*aoi+aoi,j*aoi:j*aoi+aoi,:]=img_wood_crop

plt.figure(figsize=(8,8))
plt.imshow(chess)
```
