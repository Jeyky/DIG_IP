1. Порівняти гістограми двох зображень після просторової фільтрації.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_half=cv2.imread("./assignments/Lab3_half100.png")
img_chess=cv2.imread("./assignments/Lab3_chess100.png")
```

```python
half_hist=cv2.calcHist([img_half],[0], None,[256],[0, 256])
chess_hist=cv2.calcHist([img_chess], [0], None, [256], [0,256])
```

1.3.Відобразити завантаженні зображення та відповідно визначені гістограми зображень за допомогою функції plt.subplot:

```python
plt.figure(figsize=(14,7))
plt.subplot(121)
plt.imshow(img_half)
plt.subplot(122)
plt.plot(half_hist)
plt.show()

plt.figure(figsize=(14,7))
plt.subplot(121)
plt.imshow(img_chess)
plt.subplot(122)
plt.plot(chess_hist)
plt.show()
```

1.4 Виконати просторову фільтрацію зображень використовуючи згладжувальний фільтра або розмивання Гауса з розмірами - 3x3, 5x5, 9x9, 15x15.

```python
half_blur3 = cv2.GaussianBlur(img_half,(3,3),0)
half_blur5 = cv2.GaussianBlur(img_half,(5,5),0)
half_blur9 = cv2.GaussianBlur(img_half,(9,9),0)
half_blur15 = cv2.GaussianBlur(img_half,(15,15),0)

chess_blur3 = cv2.GaussianBlur(img_chess,(3,3),0)
chess_blur5 = cv2.GaussianBlur(img_chess,(5,5),0)
chess_blur9 = cv2.GaussianBlur(img_chess,(9,9),0)
chess_blur15 = cv2.GaussianBlur(img_chess,(15,15),0)
```

1.5 Визначити гістограми для всіх отриманих зображень.

```python
half_blur3_hist = cv2.calcHist([half_blur3],[0], None, [256], [0,256])
half_blur5_hist = cv2.calcHist([half_blur5],[0], None, [256], [0,256])
half_blur9_hist = cv2.calcHist([half_blur9],[0], None, [256], [0,256])
half_blur15_hist = cv2.calcHist([half_blur15],[0], None, [256], [0,256])

chess_blur3_hist = cv2.calcHist([chess_blur3], [0], None, [256], [0,256])
chess_blur5_hist = cv2.calcHist([chess_blur5], [0], None, [256], [0,256])
chess_blur9_hist = cv2.calcHist([chess_blur9], [0], None, [256], [0,256])
chess_blur15_hist = cv2.calcHist([chess_blur15], [0], None, [256], [0,256])
```

1.6.Відобразити отримані після фільтрації зображення та відповідно визначені гістограми за допомогою функції plt.subplot:

```python
plt.figure(figsize=(18,3))
plt.subplot(121)
plt.imshow(half_blur3)
plt.subplot(122)
plt.plot(half_blur3_hist)
plt.show()
plt.figure(figsize=(18,3))
plt.subplot(121)
plt.imshow(half_blur5)
plt.subplot(122)
plt.plot(half_blur5_hist)
plt.show()
plt.figure(figsize=(18,3))
plt.subplot(121)
plt.imshow(half_blur9)
plt.subplot(122)
plt.plot(half_blur9_hist)
plt.show()
plt.figure(figsize=(18,3))
plt.subplot(121)
plt.imshow(half_blur15)
plt.subplot(122)
plt.plot(half_blur15_hist)
plt.show()
```

```python
plt.figure(figsize=(18,3))
plt.subplot(121)
plt.imshow(chess_blur3)
plt.subplot(122)
plt.plot(chess_blur3_hist)
plt.show()
plt.figure(figsize=(18,3))
plt.subplot(121)
plt.imshow(chess_blur5)
plt.subplot(122)
plt.plot(chess_blur5_hist)
plt.show()
plt.figure(figsize=(18,3))
plt.subplot(121)
plt.imshow(chess_blur9)
plt.subplot(122)
plt.plot(chess_blur9_hist)
plt.show()
plt.figure(figsize=(18,3))
plt.subplot(121)
plt.imshow(chess_blur15)
plt.subplot(122)
plt.plot(chess_blur15_hist)
plt.show()
```
