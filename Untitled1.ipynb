```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_ref=cv2.imread("./examples/Fig0304.tif",cv2.IMREAD_GRAYSCALE)
img=cv2.imread("./examples/mdb195.pgm",cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(img_ref,cmap='gray')
plt.subplot(122)
hist_ref= cv2.calcHist([img_ref],[0], None, [256], [0,256])

plt.plot(hist_ref);plt.xlim([-1, 256]); plt.grid(); plt.show();
plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(img,cmap='gray')
hist= cv2.calcHist([img], [0], None, [256], [0, 256])
plt.subplot(122)
plt.plot(hist);plt.xlim([-1,256]);plt.grid(),plt.show();
```

```python
height_ref=img_ref.shape[0]
width_ref=img_ref.shape[1]

height=img.shape[0]
width=img.shape[1]

neg_img_ref=np.zeros(img_ref.shape,dtype='uint8')
neg_img=np.zeros(img.shape,dtype='uint8')
#converting img_ref to negative
for row in range(height_ref):
    for col in range(width_ref):
        neg_ref=np.iinfo('uint8').max-img_ref[row,col]
        neg_img_ref[row,col]=neg_ref
#converting img to negative
for row in range(height):
    for col in range(width):
        neg=np.iinfo('uint8').max-img[row,col]
        neg_img[row,col]=neg

```

```python
neg_hist=cv2.calcHist([neg_img],[0],None,[256],[0,256])
neg_ref_hist=cv2.calcHist([neg_img_ref],[0],None,[256],[0,256])
plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(neg_img,cmap='gray')
plt.subplot(122)
plt.plot(neg_hist);plt.grid(),plt.show();
plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(neg_img_ref,cmap='gray')
plt.subplot(122)
plt.plot(neg_ref_hist);plt.grid();plt.show();
```

```python
neg_img_ref1=np.iinfo('uint8').max-img_ref
neg_img1=np.iinfo('uint8').max-img
plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(neg_img1,cmap='gray')
plt.subplot(122)
plt.imshow(neg_img_ref1,cmap='gray')
```

```python
#CONTRAST OPERATIONS

img_ref=cv2.imread("./examples/Fig0310b.tiff",cv2.IMREAD_GRAYSCALE)
img=cv2.imread("./examples/contr.png",cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(img_ref,cmap='gray',vmin=0,vmax=255)
plt.subplot(122)
hist = cv2.calcHist([img_ref], [0], None, [256], [0, 256])
plt.plot(hist);plt.xlim([0,256]);plt.grid();plt.show();

plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(img,cmap='gray',vmin=0,vmax=255)
plt.subplot(122)
hist=cv2.calcHist([img], [0], None, [256], [0, 256])
plt.plot(hist);plt.xlim([0,256]);plt.grid();plt.show();
```

```python
Lmin_ref=np.min(img_ref)
Lmax_ref=np.max(img_ref)

Lmin=np.min(img)
Lmax=np.max(img)

Loutmin=0
Loutmax=255

r_ref=[0,Lmin_ref,Lmax_ref,255]
s_ref=[0,Loutmax/8,Loutmax*7/8,Loutmax]

r=[0,Lmin,Lmax-30,255]
s=[0,Loutmax/14,Loutmax*3/14,Loutmax]

plt.figure(figsize=(20,10))
plt.subplot(121)
plt.plot(r_ref,s_ref,'o-')
plt.xlim([0,256]);plt.ylim([0,256]);plt.grid()

plt.subplot(122)
plt.plot(r,s,'o-')
#plt.ylabel('Інтенсивність вихідного зображення, r')
#plt.xlabel('Інтенсивність вхідного зображення, s')
plt.xlim([0,256]);plt.ylim([0,256]);plt.grid();plt.show();
```

```python
Loutmax= 255
contr_img_man_ref=np.zeros(img_ref.shape, img_ref.dtype)
alpha=int(((Loutmax*7/8)-(Loutmax/8))/(Lmax_ref-Lmin_ref))
for y in range(img_ref.shape[0]):
    for x in range(img_ref.shape[1]):
        contr_img_man_ref[y,x]=alpha*(img_ref[y,x]-Lmin_ref)
```

```python
Loutmax= 255
contr_img_man=np.zeros(img.shape, img.dtype)
alpha=int(((Loutmax*3)-(Loutmax/14))/(Lmax-Lmin))
for y in range(img.shape[0]):
    for x in range(img.shape[1]):
        contr_img_man[y,x]=alpha*(img[y,x]-Lmin)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(contr_img_man_ref, cmap='gray', vmin = 0, vmax = 255); 
plt.subplot(122)
hist = cv2.calcHist([contr_img_man_ref], [0], None, [256], [0, 256])
plt.plot(hist); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(contr_img_man, cmap='gray', vmin = 0, vmax = 255); 
plt.subplot(122)
hist = cv2.calcHist([contr_img_man], [0], None, [256], [0, 256])
plt.plot(hist); plt.xlim([0, 255]); plt.grid(); plt.show();
plt.show();

```
